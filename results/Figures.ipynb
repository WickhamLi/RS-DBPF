```python
from matplotlib import pyplot as plt
import os
import numpy as np
```

```python
mse_dpf = np.loadtxt('Poly/mse_dpf', delimiter=',', skiprows=1, usecols=(np.arange(1, 51)))
mse_pf = np.loadtxt('Poly/mse_pf', delimiter=',', skiprows=1, usecols=(np.arange(1, 51)))
mse_singdpf = np.loadtxt('Poly_single/mse_dpf', delimiter=',', skiprows=1, usecols=(np.arange(1, 51)))

# msecum_dpf = mse_dpf.cumsum(axis=1) 
# msecum_pf = mse_pf.cumsum(axis=1)
# msecum_singdpf = mse_singdpf.cumsum(axis=1)

mse_dpfmax = mse_dpf.max(axis=0)
mse_pfmax = mse_pf.max(axis=0)
mse_singdpfmax = mse_singdpf.max(axis=0)

mse_dpfmin = mse_dpf.min(axis=0)
mse_pfmin = mse_pf.min(axis=0)
mse_singdpfmin = mse_singdpf.min(axis=0)
```

```python
plt.plot(mse_dpf.mean(axis=0), label='RS-DPFs')
plt.plot(mse_pf.mean(axis=0), label='RS-PFs(Bootstrap)')
plt.plot(mse_singdpf.mean(axis=0), label='Single-DPFs')

# plt.fill_between(np.arange(50), mse_dpfmax, mse_dpfmin, alpha=0.2)
# plt.fill_between(np.arange(50), mse_pfmax, mse_pfmin, alpha=0.2)
# plt.fill_between(np.arange(50), mse_singdpfmax, mse_singdpfmin, alpha=0.2)

plt.xlabel('Time Step (Polya urn dynamics)', fontsize=14)
plt.ylabel('Average Cumulative MSE', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.yscale('log')
plt.legend(loc='lower right', fontsize=14)
plt.tight_layout()
plt.savefig('MSEcum_Poly')
plt.show()
```
